```python
import os, glob
from termcolor import colored
from colorama import Fore

class ReadData:
    def __init__(self, command, year, path):
        if command in ['-a', '-e', '-c']:
            self.command = command
            self.year = year
            self.path = path
        else:
            raise Exception("Invalid Command")
                
    def read_file(self):
        for filename in glob.glob(os.path.join(self.path, f'*{self.year}*.txt')):
            with open(os.path.join(os.getcwd(), filename), 'r') as f:
                Headings = f.readline()
                Headings = Headings.split(",")
                lines = f.readlines()
                iter = 0
                for line in lines:
                    lines[iter] = line.split(",")
                    iter += 1    
                return filename, Headings, lines
        else:
            print('The specified file does NOT exist')
    high, low, humidity = 0, 999, 0
    l_date, h_date, hu_date = None, None, None
    def max_tmp(self, highest, data, headings):
        index = headings.index("Max TemperatureC")
        GST_index = headings.index("GST")
        GST = self.h_date
        for max_index in range(0, len(data)):
            if data[max_index][index] != '':
                if int(data[max_index][index]) > int(highest):
                    highest = data[max_index][index]
                    GST = data[max_index][GST_index]
        return highest, GST
    
    def min_tmp(self, lowest, data, headings):
        index = headings.index("Min TemperatureC")
        GST_index = headings.index("GST")
        GST = self.l_date
        for max_index in range(0, len(data)):
            if data[max_index][index] != '':
                if int(data[max_index][index]) < int(lowest):
                    lowest = data[max_index][index]
                    GST = data[max_index][GST_index]
        return lowest, GST

    def max_humidity(self, highest, data, headings):
        index = headings.index(" Mean Humidity")
        GST_index = headings.index("GST")
        GST = self.hu_date
        for max_index in range(0, len(data)):
            if data[max_index][index] != '':
                if int(data[max_index][index]) > int(highest):
                    highest = data[max_index][index]
                    GST = data[max_index][GST_index]
        return highest, GST
    
    def read_files(self):
        flg = 0
        for filename in glob.glob(os.path.join(self.path, f'*{self.year}*.txt')):
            flg = 1
            with open(os.path.join(os.getcwd(), filename), 'r') as f:
                Headings = f.readline()
                Headings = Headings.split(",")
                lines = f.readlines()
                iter = 0
                for line in lines:
                    lines[iter] = line.split(",")
                    iter += 1    
                self.high, self.h_date = self.max_tmp(self.high,lines, Headings)
                self.low, self.l_date = self.min_tmp(self.low, lines, Headings)
                self.humidity, self.hu_date = self.max_humidity(self.humidity, lines, Headings)
        else:
            if flg == 0:
                print('The specified file does NOT exist')

       
class Calulation:
    def __init__(self, command, headings, data):
        self.commmand = command
        self.headings = headings
        self.data = data
        
    def max_avrg(self):
        sum = 0
        count = 0
        index = self.headings.index("Max TemperatureC")
        headings_count = len(self.headings)
        for max_index in range(0, len(self.data)):
            if self.data[max_index][index] != '':
                sum += int(self.data[max_index][index])
                count += 1
            
        return sum/count
    
    def min_avrg(self):
        sum = 0
        count = 0
        index = self.headings.index("Min TemperatureC")
        headings_count = len(self.headings)
        for max_index in range(0, len(self.data)):
            if self.data[max_index][index] != '':
                sum += int(self.data[max_index][index])
                count += 1
            
        return sum/count
    
    def humidity_avrg(self):
        sum = 0
        count = 0
        index = self.headings.index(" Mean Humidity")
        headings_count = len(self.headings)
        for max_index in range(0, len(self.data)):
            if self.data[max_index][index] != '':
                sum += int(self.data[max_index][index])
                count += 1
            
        return sum/count
    
    def bar_chart(self):
        min_index = self.headings.index("Min TemperatureC")
        max_index = self.headings.index("Max TemperatureC")
        draw_chart = lambda x : x * '+' 
        for row in range(0, len(self.data)):
            if self.data[row][max_index] == '':
                print(str(row) + draw_chart(0))
            else:
                print(str(row) + draw_chart(int(self.data[row][max_index])))
            if self.data[row][min_index] == '':
                print(str(row) + draw_chart(0))
            else:
                print(str(row) + draw_chart(int(self.data[row][min_index])))
        
command, year, file_path = input("Please provide commond, year and path").split()

Data = ReadData(command, year, file_path)

Filename, Headings, rows_data = Data.read_file()

cal = Calulation(command, Headings, rows_data)
if command == '-a':
    print(cal.max_avrg())
    print(cal.min_avrg())
    print(cal.humidity_avrg())
elif command == '-c':
    cal.bar_chart()
elif command == '-e':
    Data.read_files()
    print("Hihest {} on {}".format(Data.high,Data.h_date))
    print("Lowest {} on {}".format(Data.low,Data.l_date))    
    print("Humidity {} on {}".format(Data.humidity,Data.hu_date))    
```
